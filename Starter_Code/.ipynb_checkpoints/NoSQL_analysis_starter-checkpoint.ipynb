## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print("The number of documents containing a hygiene score of 20 is",
     establishments.count_documents(hygiene))
```

```python
# Display the first document in the results using pprint
pprint(db.establishments.find_one(hygiene))
```

```python
#Save the list of documents to a variable
hygiene_ls = list(establishments.find(hygiene))
pprint(hygiene)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hygiene_ls)

# Display the number of rows in the DataFrame
print("Rows in the hygiene_df:", len(hygiene_df))

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
Rating_Lodon_gte4 =  {'AddressLine3':{'$regex':"London"},
                     'RatingValue':{'$gte': 4}}

# Use count_documents to display the number of documents in the result
print("The number of documents containing a RatingValue greater than or equal to 4 is", 
      establishments.count_documents(Rating_Lodon_gte4))

# Display the first document in the results using pprint
pprint(db.establishments.find_one(Rating_Lodon_gte4))
```

```python
Rating_Lodon_gte4_ls = list(establishments.find(Rating_Lodon_gte4))
pprint(Rating_Lodon_gte4)
```

```python
# Convert the result to a Pandas DataFrame
Rating_Lodon_gte4_df = pd.DataFrame(Rating_Lodon_gte4_ls)

# Display the number of rows in the DataFrame
print("Rows in the Rating_Lodon_gte4_df:", len(hygiene_df))

# Display the first 10 rows of the DataFrame
Rating_Lodon_gte4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.

# Rating value must equal 5
# Sort by hygiene score

# Penang Flavours Longitude = 0.08384000 and Latitude = 51.49014200
#what you want to do is go plus or minus degree_search = 0.01

latitude = {'geocode.latitude':{'$lte':51.59, '$gte':51.39}}
longitude = {'geocode.longitude':{'$lte':0.09, '$gte':0.07}}

penang_query = {'RatingValue': 5, 
               'geocode.latitude':{'$lte':51.59, '$gte':51.39}, 
                'geocode.longitude':{'$lte':0.09, '$gte':0.07}}
sort = [("scores.Hygiene", 1)]
limit = 5

# Print the results
top5 = list(db.establishments.find(penang_query).sort(sort).limit(limit))
pprint(top5)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.DataFrame(top5)
top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
match_query = {'$match': {'scores.Hygiene': 0}}

group_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'count': {'$sum': 1}}}

sort_query = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_query]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("The number of documents in the query for 0 ratings is:", len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
zero_rating_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in the zero_rating_df:", len(zero_rating_df))

# Display the first 5 rows of the DataFrame
zero_rating_df.head(5)
```
